## helper functions

```python
from IPython.display import display_html

def _display(dfs, names=[]):
    html_str = ''
    if names:
        html_str += ('<tr>' +
                     ''.join(f'<td style="text-align:center">{name}</td>'
                             for name in names) + '</tr>')
    
    html_str += ('<tr>' + ''.join(f'<td style="vertical-align:top"> {df.to_html(index=False)}</td>'
        for df in dfs) + '</tr>')
   
    html_str = f'<table>{html_str}</table>'
    
    html_str = html_str.replace('table', 'table style="display:inline"')
    
    display_html(html_str, raw=True)
```

### Excel WorkBook for examples

```python
# Excel file to load examples from
f = 'inputs/007-MSPTDA-PowerQueryJoinsMergesStart.xlsx'
```

## Inner join

```python
def get_tables(f):
    
    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Inner', header=4)

    table1 = pd.DataFrame(tables.EmployeeNameSanJose)
    table2 = pd.DataFrame(tables.EmployeeName)
    
    return table1, table2

table1, table2 = get_tables(f)

merged_table = table1.merge(table2, how='inner', left_on='EmployeeNameSanJose', right_on='EmployeeName')

_display([table1, table2, merged_table], names=[])
```

## Full-Outer

```python
def get_tables(f):

    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Full-Outer', header=4)

    table1 = tables[['Product', 'SupplierID', 'Price', 'Cost']][:6]
    table2 = tables[['SupplierID.1', 'Name', 'City', 'State']][:3]
    
    return table1, table2

table1, table2 = get_tables(f)
table2.rename(columns={'SupplierID.1': 'SupplierID'}, inplace=True)

merged_table = table1.merge(table2, how='outer').fillna('')

_display([table1, table2, merged_table], names=[])
```

## Left-Anti

```python
def get_tables(f):

    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Left-Anti', header=4)

    table1 = pd.DataFrame(tables.EmployeeNameSanJose)
    table2 = pd.DataFrame(tables.EmployeeNameBellevue)
    
    return table1, table2
    
table1, table2 = get_tables(f)

merged_table = table1.merge(table2, how='left', 
                            left_on='EmployeeNameSanJose',
                            right_on='EmployeeNameBellevue')
merged_table = merged_table.EmployeeNameSanJose[merged_table.EmployeeNameBellevue.isna()].to_frame()

_display([table1, table2, merged_table], names=[])
```

## Right-Anti

```python
def get_tables(f):
    
    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Right-Anti', header=4)

    table1 = pd.DataFrame(tables.EmployeeNameSanJose)
    table2 = pd.DataFrame(tables.EmployeeNameBellevue)
    
    return table1, table2
    
table1, table2 = get_tables(f)    

merged_table = table1.merge(table2, how='right', 
                            left_on='EmployeeNameSanJose',
                            right_on='EmployeeNameBellevue')
merged_table = merged_table.EmployeeNameBellevue[merged_table.EmployeeNameSanJose.isna()].to_frame()

_display([table1, table2, merged_table.fillna('')], names=[])
```

## Left-Outer(1)

```python
def get_tables(f):
    
    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Left-Outer(1)', header=3)

    table1 = tables[['Product', 'Units']][:15]
    table2 = tables[['Product.1', 'Price']][:4]
    table2.rename(columns={'Product.1': 'Product'}, inplace=True)
  
    return table1, table2
    
table1, table2 = get_tables(f)    

merged_table = table1.merge(table2, how='left')

_display([table1, table2, merged_table.fillna('')], names=[])
```

## Left-Outer(2)

```python
def get_tables(f):
    

    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Left-Outer(2)', header=5)

    table1 = tables[['Product', 'Color', 'Units']][:15]
    table2 = tables[['Product.1', 'Color.1', 'Price']][:4]
    table2.rename(columns={'Product.1': 'Product', 'Color.1': 'Color'}, inplace=True)
    
    return table1, table2

table1, table2 = get_tables(f)   

merged_table = table1.merge(table2, how='left')

_display([table1, table2, merged_table.fillna('')], names=[])
```

## Left-Outer(3)

```python
def get_tables(f):

    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Left-Outer(3)', header=5)

    table1 = tables[['Invoice No.', 'Discount%']][:3]
    table2 = tables[['Invoice No..1', 'Product', 'Sales']][:6]
    table2.rename(columns={'Invoice No..1': 'Invoice No.'}, inplace=True)
    
    return table1, table2

table1, table2 = get_tables(f)   

# Aggregate the table2 data by invoice number
table2_aggregated = table2.groupby(['Invoice No.']).sum().reset_index()

merged_table = table1.merge(table2_aggregated, how='left')
merged_table['Invoice No.'] = merged_table['Invoice No.'].astype(int)

_display([table1, table2, table2_aggregated, merged_table.fillna('')], names=[])
```

## Right-Outer

```python
def get_tables(f):

    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Right-Outer', header=4)

    table1 = tables[['Product', 'SupplierID', 'Price', 'Cost']][:6]
    table2 = tables[['SupplierID.1', 'Name', 'City', 'State']][:3]
    
    return table1, table2

table1, table2 = get_tables(f)   
table2.rename(columns={'SupplierID.1': 'SupplierID'}, inplace=True)

merged_table = table1.merge(table2, how='right')
# merged_table = merged_table[~merged_table.Product.isna()]

_display([table1, table2, merged_table.fillna('')], names=[])
```

## Self-Join

```python
def get_tables(f):

    with pd.ExcelFile(f) as xlsx:
        tables = xlsx.parse(sheet_name='Self-Join', header=4)

    cols = ['EmployeeID', 'Name', 'Hire Date', 'Employee Who Referred New Employee']
    table1 = tables[cols]
    table2 = table1.copy(deep=True)
    
    return table1, table2

table1, table2 = get_tables(f)   

merged_table = table1.merge(table2, how='left', 
                            left_on='Employee Who Referred New Employee',
                            right_on='EmployeeID', suffixes=(None, '_y'))

dropped_cols = ['Employee Who Referred New Employee_y', 'EmployeeID_y', 'Hire Date_y']

merged_table = (merged_table.drop(columns=dropped_cols)
                            .rename(columns={'Name_y': 'Referral_Name'}))

_display([table1, table2, merged_table.fillna('')], names=[])
```
